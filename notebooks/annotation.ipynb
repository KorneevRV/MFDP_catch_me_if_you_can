```python
import matplotlib.pyplot as plt
```

```python
# Данные для столбцов
categories = ["0-150 т.руб\n Junior",
              "150 - 600 т.руб\n Middle / Senior / Senior+",
              "600+ т.руб\n Lead / VP / C-level" ]
values = [300, 100, 150]  # Пропорции для высоты столбцов

fig, ax = plt.subplots()

# Цвета для столбцов
colors = ['lightcoral', 'lightblue', 'gray']

# Отрисовка столбцов с индивидуальной прозрачностью
bars = ax.bar(categories, values, color=colors)
bars[2].set_alpha(0.5)

# Шриховка для третьего столбца
bars[2].set_hatch('///') 

# Подписи для столбцов белым цветом
ax.text(0, values[0] + 5, "+- 50 000 руб.", ha='center', color='white')
ax.text(1, values[1] + 5, "+- 25%", ha='center', color='white')
ax.text(2, values[2] + 5, "?", ha='center', color='white')

# Оформление графика: удаляем ось Y и рамку, фон прозрачный
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.get_yaxis().set_visible(False)  

# Устанавливаем белый цвет для подписей оси X и заголовка
ax.tick_params(axis='x', colors='white')

# Прозрачный фон
fig.patch.set_alpha(0)
ax.patch.set_alpha(0)


plt.show()
```

```python
# Сохранение графика
fig.savefig('error distribution.png', dpi=1000, transparent=True)
```
